### seaborn 데이터셋으로 연습하기 1
* sns.load_dataset('tips') : 요일별 점심, 저녁, 흡연 여부와 식사 금액과 팁을 정리한 데이터

```python
# 데이터 준비하기
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips) # x축과 y축 컬럼과 들어갈 data셋 지정
plt.show()
```

#### hue

```python
# hue라는 옵션을 이용해서 구분할 수 있다.
# hue에 smoker를 지정해줘서 흡연 여부에 따라 data가 나눠서 출력된다.
# palette - color 변경 옵션
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3')
plt.show()
```

* 흡연자가 결제 금액 범위가 비흡연자에 비해 상대적으로 크다.

```python
# swarmplot
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5")
plt.show()
```

```python
# boxplot과 swarmplot
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

#### lmplot
* data를 산포도와 같이 regression(회귀) 직선도 같이 그려주며, 유효범위도 ci(색칠된 곳)도 잡아준다!

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=10) # size 파라미터가 height로 rename 됨.
plt.show()
```

```python
# hue 파라미터도 사용 가능!
# palette로 색상 지정도 가능!
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1', height=10)
plt.show()
```

### seaborn 데이터셋으로 연습하기 2
* sns.load_dataset('flights') : 연도 및 월별 항공기 승객수를 기록한 데이터

```python
# 데이터 준비하기
flights = sns.load_dataset('flights')
flights.head()
```

```python
# 데이터 정보 확인
flights.info()
```

```python
# pivot table로 만들기
# pivot 연도별 X 월별 데이터
flights = flights.pivot(index='month', columns='year', values='passengers')
flights.head()
# pivot 기능으로 간편하게 월별, 연도별로 구분할 수 있다.
# pivot을 상상할 수 있는 것이 중요!
```

#### heatmap
* heatmap을 사용하면 이런 pivot 느낌의 데이터는 그 경향을 설명하기 참 좋다.

```python
# 히트맵 heatmap
plt.figure(figsize=(10,8))

# 파라미터
# annot=True - 각 셀에 숫자가 나타나게 함
# fmt='d' - 정수 형태(interger format)로
# cmap='YlGnBu' - 'YlGnBu' 형식으로 색상 변경
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```

```python
# 히트맵 heatmap
plt.figure(figsize=(10,8))

# 파라미터
# annot=True - 각 셀에 숫자가 나타나게 함
# fmt='d' - 정수 형태(interger format)로
# cmap='YlGnBu' - 'YlGnBu' 형식으로 색상 변경
sns.heatmap(flights, annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
US_Unemployment_Oct2012 = f'{url}/US_Unemployment_Oct2012.csv'
unemployment = pd.read_csv(US_Unemployment_Oct2012)

unemployment.head(5)


from branca.colormap import linear

colormap = linear.YlGn_09.scale(
    unemployment.Unemployment.min(),
    unemployment.Unemployment.max())

print(colormap(5.0))

colormap
```

```python
unemployment.head()
```

```python
unemployment_dict = unemployment.set_index('State')['Unemployment']

unemployment_dict['AL']
```

```python
unemployment_dict
```

```python
m = folium.Map([43, -100], zoom_start=4)

folium.GeoJson(
    geo_json_data,
    name='unemployment',
    style_function=lambda feature: {
        'fillColor': colormap(unemployment_dict[feature['id']]),
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5',
        'fillOpacity': 0.9,
    }
).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
color_dict = {key: colormap(unemployment_dict[key]) for key in unemployment_dict.keys()}
```

```python
m = folium.Map([43, -100], zoom_start=4)

folium.GeoJson(
    geo_json_data,
    style_function=lambda feature: {
        'fillColor': color_dict[feature['id']],
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5',
        'fillOpacity': 0.9,
    }
).add_to(m)
```
