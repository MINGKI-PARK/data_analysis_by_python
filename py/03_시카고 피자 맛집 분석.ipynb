### 정규식 사용하기

```python
import re
```

```python
tmp_string = tmp_one.find(class_='sammyListing').get_text()
tmp_string
```

```python
re.split(('\n|\r\n'), tmp_string)

print(re.split(('\n|\r\n'), tmp_string)[0])
print(re.split(('\n|\r\n'), tmp_string)[1])
```

```python
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    
    # 순위
    rank.append(item.find(class_='sammyRank').get_text())
    
    # 메뉴, 카페 이름
    tmp_string = item.find(class_='sammyListing').get_text()
    
    # 메뉴
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    
    # 카페 이름
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    # url
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

```python
main_menu[:5]
```

```python
cafe_name[:5]
```

```python
url_add[:5]
```

```python
len(rank), len(main_menu), len(cafe_name), len(url_add)
```

```python
import pandas as pd
```

```python
data = {'Rank' : rank,
        'Menu' : main_menu,
        'Cafe' : cafe_name,
        'URL' : url_add}
df = pd.DataFrame(data)
df.head()
```

```python
# 컬럼의 순서 재정렬하기
df = pd.DataFrame(data, columns=['Rank', 'Cafe', 'Menu', 'URL'])
df.head()
```

```python
# 저장하기
df.to_csv('../data/03_best_sandwiches_list_chicago.csv', sep=',', encoding='UTF-8')
```

## 맛집 위치를 지도에 표기하기

```python
import numpy as np
import pandas as pd
import folium
import googlemaps
```

```python
df = pd.read_csv('../data/03_best_sandwiches_list_chicago2.csv', index_col=0)
df.head()
```

```python
gmaps_key = 'AIzaSyBnGW4gT4pQ7bzEiLxx0fIYMv1s0AWB2Bc'
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
gmaps.geocode('2109 W. Chicago Ave., Chicago')
```

```python
lat = []
lng = []

for n in tqdm.notebook.tqdm(df.index):
    if df['Address'][n] != 'Multiple': # 주소에 Multiple이 나타나지 않은 경우만 주소를 검색하도록 한다.
        target_name = df['Address'][n] + 'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
df['lat'] = lat
df['lng'] = lng
```

```python
df.head()
```

```python
# 지도의 중앙값 찾기
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                     zoom_start=11)

folium.Marker([df['lat'].mean(), df['lng'].mean()],
              popup='center').add_to(mapping)

mapping
```

```python
# 가게들의 중앙값으로 지도 위치 잡기
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)

# 지도에 가게 표시하기
for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]], popup=df['Cafe'][n]).add_to(mapping)
        
mapping
```

### 접근한 웹 페이지에서 원하는 데이터 추출하고 정리하기

```python
tmp_one = soup.find_all('div', 'pizza15-list-item')[0]
type(tmp_one) # bs4.element.Tag라고 하는 형태로 이런 경우 그 변수에 다시 태그로 찾는 find or find_all 명령을 사용할 수 있다.
```

```python
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'pizza15-list-item')

for item in list_soup:
    
    # 랭킹
    rank.append(item.find(class_='pizza15-rank').get_text())
    
    # 가게 이름
    cafe_name.append(item.find('a').get_text())
    
    # url
    url_add.append(url_base + item.find('a')['href'])
    
    # 메뉴
    main_menu.append(' '.join(item.find_all('li')[3].get_text().split()[2:]))
```
