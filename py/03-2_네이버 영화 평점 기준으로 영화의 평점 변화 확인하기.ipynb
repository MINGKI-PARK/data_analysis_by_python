```python
import numpy as np
import pandas as pd

import tqdm

from bs4 import BeautifulSoup
from urllib.request import urlopen, urljoin

from urllib import parse
```

## 영화의 평점 변화 확인하기

```python
# url 주소 가져오기
url_base = 'https://movie.naver.com'
url_sub = '/movie/sdb/rank/rmovie.nhn?sel=cur&date=20201007'

page = urlopen(url_base+url_sub)

soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
# 태그 가져오기
soup.find_all('td', 'title')
```

```python
# 제목만 찾기
soup.find_all('td', 'title')[0].a.string
```

```python
# 평점 찾기
soup.find_all('td', 'point')[0].string
```

```python
# 날짜 지정해 주기
date = pd.date_range('2020-7-1', periods=100, freq='D')
date
```

```python
# 날짜 지정해서 정보 가져오기

movie_date = []
movie_name = []
movie_point = []

for today in tqdm.notebook.tqdm(date):
    
    # 주소 설정
    html = url_base + '/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}'.format(date=parse.quote(today.strftime('%Y%m%d')))
    
    response = urlopen(html)
    
    soup = BeautifulSoup(response, 'html.parser')
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)]) # extend를 사용한 이유!
    movie_name.extend([soup.find_all('td', 'title')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
# 데이터 확인
len(movie_date), len(movie_name), len(movie_point)
```

```python
# 데이터프레임으로 만들기
movie = pd.DataFrame({'date' : movie_date,
                      'name' : movie_name,
                      'point' : movie_point})

movie.head()
```

```python
movie.info()
```

```python
# 영화 평점의 데이터 타입 변경해 주기
movie['point'] = movie['point'].astype(float)

movie.info()
```

```python
# 피봇 테이블로 변경해서 평점 고득점 영화 파악하기
movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
# 특정 영화만 추려서 확인하기
tmp = movie.query('name == ["소년시절의 너"]')
tmp
```

```python
# 날짜별 평점 변화 시각화 해보기
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'], label='point')
plt.legend()
plt.grid()
plt.show()
```

## 영화별 날짜 변화에 따른 평점 변화 확인하기

```python
movie_pivot = pd.pivot_table(movie, index=['date'], columns='name', values=['point'])

movie_pivot.head()
```

```python
# 영화 제목 위에 붙은 포인트 컬럼 제거하기
movie_pivot.columns = movie_pivot.columns.droplevel()
movie_pivot.head()
```
